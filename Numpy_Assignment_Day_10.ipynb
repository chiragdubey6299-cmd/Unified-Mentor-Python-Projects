## Numpy copy vs view

```python

x1 = np.array([10, 20, 30, 40, 50])

print("Original array x1:", x1)

```

```python

x2 = x1.view()

print("View array x2:", x2)

```

```python

print("x1:", x1)
print("x2:", x2)


```

Ok now you have seen that both of them are same 

```python

x2[0] = 99
print("x2 changed to:", x2)


```

```python

print("x1 (after change in x2):", x1)
print("x2 (after change in x2):", x2)

```

```python

print("x1.base is:", x1.base)
print("x2.base is:", x2.base)
print("Is x2 a view of x1? (x2.base is x1):", x2.base is x1)

```

Hey It's True they both share memory

Shall we try **view()** function also likwise.

```python

x3 = x1.view()

print("x3 (View of x1):", x3)
```

```python

x3[0] = 999
print("x1 (after change in x3):", x1)
print("x3 (after change in x3):", x3)


```

Woh! simple assignment is similar to view. That means 
The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view.

Don't agree? ok lets change x3 and see if original array i.e. x1 also changes

```python

x3[0] = 100

print("x3 updated:", x3)

```

```python

print("x1 (Original array after change in x3):", x1)
print("x3 (View array):", x3)


```

Now its proved.

Lets see how **Copy()** function works

```python

x4 = x1.copy()

print("x4 (Copy of x1):", x4)

```

```python

x4[-1] = 999

print("x4 updated:", x4)



```

```python

print("x1 (Original array after change in x4):", x1)
print("x4 (Copy array):", x4)



```

Hey! such an intresting output. You noticed buddy! your original array didn't get changed on change of its copy ie. x4.

Still not convinced? Ok lets see if they both share memory or not

```python

print("x1.base is:", x1.base)
print("x4.base is:", x4.base)
print("Are x1 and x4 the same object? (x1 is x4):", x1 is x4)


```

**hstack vs vstack function**


Stacking is same as concatenation, the only difference is that stacking is done along a new axis.

NumPy provides a helper function: 

1. hstack() to stack along rows.
2. vstack()  to stack along columns

```python

print(np.hstack((x1, x4)))

```

```python

print(np.vstack((x1, x4)))

```

We hope now you saw the difference between them.

Fun fact! you can even use concatenate() function to join 2 arrays along with the axis. If axis is not explicitly passed, it is taken as 0 ie. along column

Lets try this function as well

```python

import numpy as np 
arr1 = np.array([[1, 2], 
                 [3, 4]])
arr2 = np.array([[5, 6], 
                 [7, 8]])

print("arr1:\n", arr1)
print("arr2:\n", arr2)

```

```python

result_rows = np.concatenate((arr1, arr2), axis=0)
print("\nConcatenate along rows (axis=0):\n", result_rows)

```
